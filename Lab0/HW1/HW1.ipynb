ابتدا هیستوگرام امتیاز نهایی تیم ها در دو فصل ۲۰۰۰ و ۲۰۱۶ را همزمان رسم کنید
.دقت کنید در رسم هیستوگرام محدوده امتیازهای مورد استفاده بین ماکزیمم و مینیمم مجموع داده های دو فصل است
</div>

```python
# filter a part of dataframe that is for season 2000-01
laliga_2000 = laliga_df[laliga_df['season'] == "2000-01"]

# filter a part of dataframe that is for season 2016-17
laliga_2016 = laliga_df[laliga_df['season'] == "2016-17"]


# Combine both datasets to calculate the bin edges
combined_data = pd.concat([laliga_2000["points"], laliga_2016["points"]])

# Calculate the minimum and maximum values in the combined data
min_value = combined_data.min()
max_value = combined_data.max()

# Number of bins
num_bins = 9

# Calculate the bin width
bin_width = (max_value - min_value) / num_bins

# Create an array of bin edges that spans from the min_value to max_value
bin_edges = [min_value + i * bin_width for i in range(num_bins + 1)]

# plotting
plt.figure(figsize=(10, 6))
plt.hist(laliga_2000["points"], bins=bin_edges, color='c', alpha=0.7, label='2000-01')
plt.hist(laliga_2016["points"], bins=bin_edges, color='m', alpha=0.7, label='2016-17')
plt.xlabel('Points')
plt.ylabel('Number of Teams')
plt.title('Histogram of Points: 2000 vs 2016')
plt.legend()
```


<div dir=rtl>

در گام بعد هیستوگرام تمام فصل های بین سال ۲۰۰۰ تا ۲۰۱۶ را جداگانه رسم و داده های آن ها را ذخیره کنید. محدوده امتیازها و دسته ها در تمام هیستوگرام ها یکسان و بین ماکزیمم و مینیمم تمام داده های تمام فصل ها است
.آیا روند خاصی در گذر زمان مشاهده میکنید؟
</div>


```python
# required colors for plotting
colors = ['red', 'blue', 'green', 'purple', 'orange', 'brown', 'pink', 'gray', 'cyan', 'magenta', 'lime', 'teal', 'olive', 'navy', 'maroon', 'gold', 'indigo']

# minimum points of all seasons
min_score = laliga_df['points'].min()

# maximum points of all seasons
max_score = laliga_df['points'].max()

# number of bins
num_bins = 20

# filter data from season 2000-01 to 2016-17
filtered_data = laliga_df[(laliga_df['season'] >= '2000-01') & (laliga_df['season'] <= '2016-17')]

# generate figure for plot histograms
fig, axes = plt.subplots(nrows=5, ncols=4, figsize=(16, 16))
plt.subplots_adjust(hspace=0.5)

# loop for plot histogram of filtered seasons
for i, (season, group_data) in enumerate(filtered_data.groupby('season')):
    # define location of histogram
    row = i // 4
    col = i % 4
    
    # plot histogram
    axes[row, col].hist(group_data['points'], bins=num_bins, range=(min_score, max_score), color=colors[i])
    axes[row, col].set_title(f'season {season}')
    axes[row, col].set_xlabel('Points')
    axes[row, col].set_ylabel('Number of Teams')

# delete extra subplots
for i in range(len(filtered_data['season'].unique()), 20):
    fig.delaxes(axes[i // 4, i % 4])

# save as image
plt.savefig('histograms_of_2000-2016.png')
plt.show()
```

<div dir=rtl>

این معیار را بین هر دو هیستوگرام موجود به صورت دستی محاسبه کرده و نتایج را در یک جدول ۱۶ * ۱۶ ارائه دهید. برداشت خود از نتایج را توضیح دهید
</div>

```python
# function that calculate Chi-Square distance
def chi_square_distance(hist1, hist2):
    distance = 0.0
    for x, y in zip(hist1, hist2):
        if x + y == 0:
            continue
        distance += ((x - y) ** 2) / (x + y)
    
    distance *= 0.5
    return distance

# number of filtered seasons
num_seasons = len(filtered_data['season'].unique())

# create table 
chi_square_table = np.zeros((num_seasons, num_seasons))

# converting histograms to numeric vectors
histograms = []
for season, group_data in filtered_data.groupby('season'):
    hist, _ = np.histogram(group_data['points'], bins=num_bins, range=(min_score, max_score))
    histograms.append(hist)

# calculate Chi-Square distance for different histograms
for i in range(num_seasons):
    for j in range(i, num_seasons):
        chi2= chi_square_distance(histograms[i], histograms[j])
        chi_square_table[i, j] = chi2
        chi_square_table[j, i] = chi2

# convert table to dataframe
chi_square_df = pd.DataFrame(chi_square_table, columns=filtered_data['season'].unique(), index=filtered_data['season'].unique())

# print dataframe
print(chi_square_df)
```

<div dir=rtl>
<span style="color: blue;">
این فاصله یک معیار آماری است که برای مقایسه شباهت یا عدم تشابه بین دو توزیع احتمال یا هیستوگرام استفاده می شود.
عناصر روی قطر اصلی نشان دهنده مقایسه هر فصل با خودش است، بنابراین مقادیر در این سلول ها همه صفر هستند. این به این دلیل است که این معیار بین یک توزیع یکسان صفر است که نشان دهنده تطابق کامل است.
مقادیر جدول متقارن هستند، به این معنی که این فاصله بین فصل A و فصل B با فاصله بین فصل B و فصل A یکسان است. این یکی از ویژگی های متریک این فاصله است.
باقی عناصر نشان دهنده مقایسه فصول مختلف است. مقادیر موجود در این سلول ها نشان می دهد که هیستوگرام ها چقدر بین فصول مربوطه متفاوت هستند. مقادیر بالاتر نشان دهنده عدم تشابه بیشتر است، در حالی که مقادیر کمتر نشان دهنده شباهت بیشتر است.
</span>

<div dir=rtl>

دقت کنید که ممکن است ماکزیمم( یا مینیمم) امتیازها در یک سال برای دو تیم برابر باشد. در این حالت تیم با تفاضل گل بهتر (یا بدتر) به عنوان خروجی تابع
در نظر گرفته میشود
</div>

```python
def min_points(df_laliga,season):
    # filter data of desired season
    season_data = df_laliga[df_laliga['season'] == season]
    # get teams with minimum points
    min_points_team = season_data[season_data['points'] == season_data['points'].min()]
    # comparison of goal difference in the case of equality of points
    if len(min_points_team) > 1:
        min_points_team = min_points_team[min_points_team['goal_difference'] == min_points_team['goal_difference'].min()]
    
    min_team_info = {
        'club': min_points_team['club'].iloc[0],
        'points': min_points_team['points'].iloc[0]
    }
    
    return min_team_info
```

```python
def max_points(df_laliga,season):
    # filter data of desired season
    season_data = df_laliga[df_laliga['season'] == season]
     # get teams with maximum points
    max_points_team = season_data[season_data['points'] == season_data['points'].max()]
    # comparison of goal difference in the case of equality of points
    if len(max_points_team) > 1:
        max_points_team = max_points_team[max_points_team['goal_difference'] == max_points_team['goal_difference'].max()]
    
    max_team_info = {
        'club': max_points_team['club'].iloc[0],
        'points': max_points_team['points'].iloc[0]
    }
    
    return max_team_info
```

حال از شما می‌خواهیم تابعی بنویسید تا با گرفتن نام فصل جدول آن فصل را خروجی دهد.

 دقت کنید تیم‌ها در جدول یک فصل باید به ترتیب امتیاز و سپس تفاضل گل مرتب شده باشند.

 دقت کنید`DataFrame` لالیگا به عنوان آرگومان اول به تابع ورودی داده می‌شود.
 </div>

```python
def get_season_table(df_laliga,season_name):
    # filter data of desired season
    season_data = df_laliga[df_laliga['season'] == season_name]
    # sort table in order of points and then goal difference
    sorted_season_table = season_data.sort_values(by=['points', 'goal_difference'], ascending=[False, False])
    
    return sorted_season_table
```

سلول زیر را اجرا کنید تا فایل های csv  را به همراه این نوت‌بوک زیپ کند.

```python
import zipfile
import os


def compress(file_names):
    print("File Paths:")
    print(file_names)

    # Select the compression mode ZIP_DEFLATED for compression
    # or zipfile.ZIP_STORED to just store the file
    compression = zipfile.ZIP_DEFLATED

    # create the zip file first parameter path/name, second mode
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            # Add file to the zip file
            # first parameter file to zip, second filename in zip
            zf.write('./' + file_name, file_name, compress_type=compression)


file_names = [
    file for file in os.listdir()
    if (file.endswith('csv') or file.endswith('ipynb'))
    and file != 'LaLiga_dataset.csv'
]

compress(file_names)
```
